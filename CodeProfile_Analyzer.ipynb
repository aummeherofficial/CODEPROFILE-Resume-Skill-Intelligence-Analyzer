```python
print("\nPaste your Full Resume Text Below : \n")
resume_text = input()
print("Resume Received Successfully")
```

```python
skill_database = {

    "Programming Languages" : {
        "Python", "java", "c", "c++", "javascript", "php", "kotlin"
    },

    "Web Development" : {
        "html", "css", "react", "vue", "angular", "flask", "node", "express", "django"
    },

    "Data Science" : {
        "pandas", "numpy", "matplotlib", "seaborn", "scikit-learn", "statistics", "data analysis", "data analytics"
    },

    "AI / Machine Learning" : {
        "machine learning", "deep learning", "tensorflow", "keras", "nlp", "computer vision"
    },

    "Database" : {
        "oracle", "sql", "mysql", "sqlite", "postgresql", "mongodb"
    },

    "Cloud & DevOps": {
        "aws", "azure", "gcp", "docker", "kubernetes", "ci/cd"
    },

    "Tools" : {
        "git", "github", "excel", "power bi", "tableau", "jira"
    }

}
```

###CONVERT THE RESUME TEXT TO LOWERCASE

```python
resume_text = resume_text.lower()
print("Resume Text Converted to Lowercase")
```

###SKILL DETECTION LOGIC

```python
# Dictionary to store detected skills
detected_skills = {}

# Loop through each category
for category, skills in skill_database.items():
 # Create empty set to store matched skills
  matched = set()
  for skill in skills:
    if skill in resume_text:
      matched.add(skill)

  # Store only if at least one skill found
  if matched:
    detected_skills[category] = matched
```

###DISPLAY DETECTED SKILLS

```python
print("\n=============DETECTED SKILLS=============\n")
if not detected_skills:
  print("No IT Skill Detected")
else:
  for category, skills in detected_skills.items():
    print(f"{category}")
    for skill in skills:
      print(" -", skill)
    print()
```

###Profile Strength Scoring System

```python
total_skills = 0
total_categories = len(detected_skills)

#count total skills

for skills in detected_skills.values():
  total_skills += len(skills)

#scoring logic
score = total_skills * 10

# Bonus if multiple categories covered
if total_categories >= 3:
  score += 20

# Bonus if AI/ML present
if "Data Science" in detected_skills:
  score += 10
```

```python
# Cap score at 100
if score > 100:
    score = 100

print("\n=============PROFILE SCORE=============\n")
print("Total Skills Detected :", total_skills)
print("Total Categories Covered :", total_categories)
print("Final Resume Score :", score, "/ 100")


# Profile Level Classification
if score < 30:
    level = "Beginner Profile"
elif score < 60:
    level = "Intermediate Profile"
elif score < 85:
    level = "Strong Profile"
else:
    level = "Excellent / Industry Ready"

print("Profile Level :", level)


# ==========================
# Skill Improvement Suggestions
# ==========================

print("\n=============IMPROVEMENT SUGGESTIONS=============\n")

missing_categories = []

for category in skill_database:
    if category not in detected_skills:
        missing_categories.append(category)

if not missing_categories:
    print("Amazing! You have skills in all major categories 🔥")
else:
    print("You can improve your profile by adding skills in:")
    for category in missing_categories:
        print(" -", category)
```
